```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
titanic = pd.read_csv(data_url)
```

```python
# Basic exploration
titanic.head()
```

```python
titanic.shape
```

```python
titanic.isnull().sum()
```

```python
# Handle Missing Values
age_mean = titanic['Age'].mean()
titanic['Age'] = titanic['Age'].fillna(age_mean)
```

```python
# Replace missing values in 'Grade' with the mode
Cabin_mode = titanic['Cabin'].mode()[0]
titanic['Cabin'] = titanic['Cabin'].fillna('cabin_mode')
```

```python
Cabin_mode
```

```python
titanic['Embarked'] = titanic['Embarked'].fillna(titanic['Embarked'].mode()[0])
```

```python
titanic.isnull().sum()
```

```python
# Histogram for 'Age' and 'Fare'
plt.hist(titanic['Age'], bins=20, color='blue', alpha=0.7)
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

plt.hist(titanic['Fare'],bins=20, color='red', alpha=0.7)
plt.title('Distribution of Fare')
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.show()
```

```python
# Correlation analysis for numerical features
numerical_cols = titanic.select_dtypes(include=['float64', 'int64']).columns
corr_matrix = titanic[numerical_cols].corr()

sns.heatmap(corr_matrix, annot=True)
plt.title('Correlation Matrix')
plt.show()
```

```python
# Scatter plot for Age vs. Fare
plt.scatter(titanic['Age'], titanic['Fare'], color='blue', alpha=0.5)
plt.title('Scatter Plot: Age vs Fare')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```

```python
titanic.set_index('PassengerId',inplace=True)
```

```python
# Scatter plot for Age vs. Fare
plt.scatter(titanic.index,titanic['Age'], color='blue', alpha=0.5)
plt.show()
```

```python
survival = titanic.groupby('Sex')['Survived'].sum()

# Plot the bar chart
plt.bar(survival.index,survival.values, color=['blue', 'orange'])

# Labels and title
plt.title('Total Survivors by Gender')
plt.xlabel('Gender')
plt.ylabel('Number of Survivors')

# Show the plot
plt.show()
```

```python
survival
```

```python
plt.pie(survival.values,labels=survival.index, autopct='%1.1f%%',colors=['blue', 'orange','Red'])
plt.title('Total Passanger Survived')
# Show plot
plt.show()
```
